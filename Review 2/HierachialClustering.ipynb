# Aglometrative Clustering
Hierarchical Agglomerative Clustering (HAC) is a type of clustering algorithm that is used to cluster a set of data points into groups (also known as clusters) based on their similarity. HAC is a bottom-up approach, in which each data point is treated as a single cluster at the outset and the algorithm iteratively merges the most similar clusters until a specified stopping condition is reached.

```python
#Using to HAC to form clusters for a given number of clusters
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))
#we are essentially stiching both the x & y attributes into a single set of points(x,y) 

linkage_data = linkage(data, method='complete', metric='euclidean')
#This function takes 3 parameters: dataset, linkage_method, distance_measure
dendrogram(linkage_data)

plt.show()
```
